```python
import json

def load(filename):
    with open(filename,'r') as f:
        return json.load(f)

def code(data,userid):
    pagev={}
    for p in data['users']:
        pagev[p['id']]=set(p['liked_pages'])
    if userid not in pagev:
        return[]
    ulp=pagev[userid]

    pgsuggest={}
    for u,p in pagev.items():
        if u!= userid:
            Intersection=ulp.intersection(p)
            for sp in p:
                if p not in ulp:
                    pgsuggest[sp]=pgsuggest.get(sp,0)+len(Intersection)
    short=sorted(pgsuggest.items(),key = lambda x:x[1], reverse= True)
    return[f  for f in short]
    
        

data=load('new.json')
userid=2
done=code(data,userid)
print(done)
```
